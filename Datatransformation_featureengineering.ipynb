Checking for missing values in the dataset.

# Missing Value Treatment

Before feature engineering, missing values must be handled to ensure data quality.

### Numerical Columns
Missing values are replaced using the median value.

### Categorical Columns
Missing values are replaced using the most frequent category (mode).

This approach preserves the dataset while minimizing the impact of missing data.

```python
# Separate numerical and categorical columns

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Fill numerical missing values with median

for col in numerical_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill categorical missing values with mode

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Check again

df.isnull().sum().sum()
```

# Missing Value Treatment Results

After applying median imputation for numerical features and mode imputation for categorical features, all missing values were successfully removed from the dataset.

This ensures that the dataset is complete and ready for feature engineering techniques such as One-Hot Encoding and Feature Scaling.

# One-Hot Encoding

Many machine learning algorithms cannot work directly with categorical variables.

One-Hot Encoding converts categorical features into numerical binary columns.

This transformation allows machine learning models to process categorical information effectively.

```python
categorical_cols = df.select_dtypes(include=['object']).columns

print("Number of categorical columns:", len(categorical_cols))
```

```python
df_encoded = pd.get_dummies(
    df,
    columns=categorical_cols,
    drop_first=True
)

print("Original Shape:", df.shape)
print("Encoded Shape:", df_encoded.shape)
```

# One-Hot Encoding Results

After encoding, categorical features were transformed into binary variables.

This increased the number of columns in the dataset because each category was converted into a separate feature.

# Feature Scaling

Feature scaling is used to standardize numerical features.

StandardScaler transforms data so that:
- Mean = 0
- Standard Deviation = 1

This prevents features with larger values from dominating machine learning algorithms.

```python
from sklearn.preprocessing import StandardScaler

numerical_cols = df_encoded.select_dtypes(
    include=['int64', 'float64']
).columns

scaler = StandardScaler()

df_encoded[numerical_cols] = scaler.fit_transform(
    df_encoded[numerical_cols]
)

df_encoded.head()
```

# Feature Scaling Results

Numerical features were standardized using StandardScaler.

The transformed features now have a consistent scale, which improves model performance and training stability.

# Creating Interaction Features

Interaction features are generated by combining existing features.

These features help machine learning models capture relationships between variables that may not be apparent when using individual features alone.

```python
df_encoded["Quality_Area_Interaction"] = (
    df["OverallQual"] * df["GrLivArea"]
)

df_encoded[
    [
        "OverallQual",
        "GrLivArea",
        "Quality_Area_Interaction"
    ]
].head()
```

# Interaction Feature Results

A new feature named Quality_Area_Interaction was created by multiplying Overall Quality and Ground Living Area.

This feature represents the combined effect of house quality and living area on property value.
